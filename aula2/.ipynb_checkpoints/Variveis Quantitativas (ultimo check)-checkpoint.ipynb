```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
from numpy import arange

#Abrindo Arquivo
dados = pd.read_table("pes2013A.txt", encoding="windows-1252", sep=" ")

#Criando DataFrame
escolhas = pd.DataFrame()

escolhas["Unidade da federação"] = dados.UF
escolhas["Regiões"] = dados.UF
escolhas["Idade"] = dados.V8005
escolhas["Rend"] = dados.V4720
escolhas["Académico"] = dados.V8005

#rendatotal = escolhas[(escolhas.Rend != 999999999999)].Rend

n = len(escolhas)
escolhas.Regiões = pd.Series([""]*n)

#Redividindo Regiões
escolhas.Regiões[(dados.UF>=11) & (dados.UF<=17)] = "Norte"
escolhas.Regiões[(dados.UF>=21) & (dados.UF<=29)] = "Nordeste"
escolhas.Regiões[(dados.UF>=31) & (dados.UF<=35)] = "Sudeste"
escolhas.Regiões[(dados.UF>=41) & (dados.UF<=43)] = "Sul"
escolhas.Regiões[(dados.UF>=50) & (dados.UF<=53)] = "Centro-Oeste"

escolhas.Académico = pd.Series([""]*n)
escolhas.Académico[(dados.V8005 >=17) & (dados.V8005 <30)] = "Académico"
```

Criando Histogramas

```python
#Idade

escolhas.Idade.describe()

faixa = arange(0,120,10)

plot = escolhas.Idade.plot.hist(bins=faixa, normed=1, color="red")
plt.title("Histograma de Idade")
plt.xlabel("Idade (Anos)")
plt.ylabel("Frequência relativa")
```

```python
#Renda

escolhas.Rend.describe()

faixa2 = arange(0,6000,200)


plot = escolhas.Rend.plot.hist(bins=faixa2, normed=1, color="green")
plt.title("Histograma de Renda")
plt.xlabel("Renda Mensal (reais)")
plt.ylabel("Frequência relativa")
```

```python
#Análise por Região - Histogramas

faixa = arange(0,120,10)
faixa2 = arange(0,6000,200)

#Norte
idadenorte = escolhas[(escolhas.Regiões == "Norte")].Idade

plot = idadenorte.plot.hist(bins=faixa, color="red")
plt.title("Idade Região Norte")
plt.xlabel("Idade (Anos)")
plt.ylabel("Frequência relativa")
plt.show()

rendanorte = escolhas[(escolhas.Regiões == "Norte")].Rend

plot = rendanorte.plot.hist(bins=faixa2, color="green")
plt.title("Renda mensal na Região Norte")
plt.xlabel("Renda (Reais)")
plt.ylabel("Frequência relativa")
plt.show()

#Nordeste
idadenordeste = escolhas[(escolhas.Regiões == "Nordeste")].Idade

plot = idadenordeste.plot.hist(bins=faixa, normed=1, color="red")
plt.title("Idade Região Nordeste")
plt.xlabel("Idade (Anos)")
plt.ylabel("Frequência relativa")
plt.show()

rendanordeste = escolhas[(escolhas.Regiões == "Nordeste")].Rend

plot = rendanordeste.plot.hist(bins=faixa2, normed=1, color="green")
plt.title("Renda mensal na Região Nordeste")
plt.xlabel("Renda (Reais)")
plt.ylabel("Frequência relativa")
plt.show()

#Sudeste
idadesudeste = escolhas[(escolhas.Regiões == "Sudeste")].Idade

plot = idadesudeste.plot.hist(bins=faixa, normed=1, color="red")
plt.title("Idade Região Sudeste")
plt.xlabel("Idade (Anos)")
plt.ylabel("Frequência relativa")
plt.show()

rendasudeste = escolhas[(escolhas.Regiões == "Sudeste")].Rend

plot = rendasudeste.plot.hist(bins=faixa2, normed=1, color="green")
plt.title("Renda mensal na Região Sudeste")
plt.xlabel("Renda (Reais)")
plt.ylabel("Frequência relativa")
plt.show()

#Sul
idadesul = escolhas[(escolhas.Regiões == "Sul")].Idade

plot = idadesul.plot.hist(bins=faixa, normed=1, color="red")
plt.title("Idade Região Sul")
plt.xlabel("Idade (Anos)")
plt.ylabel("Frequência relativa")
plt.show()

rendasul = escolhas[(escolhas.Regiões == "Sul")].Rend

plot = rendasul.plot.hist(bins=faixa2, normed=1, color="green")
plt.title("Renda mensal na Região Nordeste")
plt.xlabel("Renda (Reais)")
plt.ylabel("Frequência relativa")
plt.show()

#Centro-Oeste
idadecentrooeste = escolhas[(escolhas.Regiões == "Centro-Oeste")].Idade

plot = idadecentrooeste.plot.hist(bins=faixa, normed=1, color="red")
plt.title("Idade Região Centro-Oeste")
plt.xlabel("Idade (Anos)")
plt.ylabel("Frequência relativa")
plt.show()

rendacentrooeste = escolhas[(escolhas.Regiões == "Centro-Oeste")].Rend

plot = rendacentrooeste.plot.hist(bins=faixa2, normed=1, color="green")
plt.title("Renda mensal na Região Centro-Oeste")
plt.xlabel("Renda (Reais)")
plt.ylabel("Frequência relativa")
plt.show()
```

Podemos observar que a região Norte apresenta grande quantidade de habitantes com idade relativamente académica (entre 17 e 30 anos), além de apresentar uma faixa de 100 mil habitantes com renda per cápita entre 600 e 1400 reais.
